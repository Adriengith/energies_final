```python
import os
import csv
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup

page = "https://www.la-spa.fr/adopter-animaux?page="
filename = "spa.csv"
f = open(filename, "w")

headers = "name, refuge_name\n"
f.write(headers)

for page_number in range ( 1,223):
    my_url = page + str(page_number)
    print(my_url)
    uClient = uReq(my_url)
    page_html = uClient.read()
    uClient.close()

    page_soup = soup(page_html, "html.parser")

    containers = page_soup.findAll("div",{"class":"block-result-search-bottom"})

    for container in containers:
        name = container.a.text
        refuge_name = container.find("span", {"class":"refuge-name"}).a.text

    print("name: " + name)
    print("refuge_name: " + refuge_name)

    f.write(name + "," + refuge_name + "\n")
f.close()
```
